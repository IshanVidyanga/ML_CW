# Encoding data before applying ML algorithm

```python
from sklearn.preprocessing import LabelEncoder
lblEnc = LabelEncoder()
df_churn = df.copy()

for i in df.columns :
    df_churn[i] = lblEnc.fit_transform(df[i])
```

```python
df_churn
```

```python
X = df_churn.drop('spam', axis = 1)
Y = df_churn['spam']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
#30% data is used for testing and 80% for training

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.30, random_state = 2)
```

```python
# Perform feature scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train.shape, '----> Training Feature Shape')
print(X_test.shape, '----> Training Feature Shape')
print(Y_train.shape, '----> Testing Labels Shape')
print(Y_test.shape, '----> Testing Feature Shape')
```

# Building KNN model


```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

# Perform grid search to find optimal value of k
param_grid = {'n_neighbors': [1, 3, 5, 7, 9]}
grid_search = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5)
grid_search.fit(X_train, Y_train)

# Print optimal value of k
print('Optimal value of k:', grid_search.best_params_['n_neighbors'])
```

Using k = 5 as the optimal value for the knn model.We will train model by 80% of the training set and evaluate the models performance by the 20% of training data set.

We will use k=5 as the optimal value for our KNN model. We will train the model on the training set and evaluate its performance on the testing set.

# Training the KNN model
k = 5

```python
# Build KNN model with k=5
knn = KNeighborsClassifier(n_neighbors=5)

# Train the model
knn.fit(X_train, Y_train)

# Evaluate the model on testing set
Y_pred = knn.predict(X_test) 

# Print evaluation metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print('Accuracy:', accuracy_score(Y_test, Y_pred))
print('Precision:', precision_score(Y_test, Y_pred))
print('Recall:', recall_score(Y_test, Y_pred))
print('F1 Score:', f1_score(Y_test, Y_pred))
```
